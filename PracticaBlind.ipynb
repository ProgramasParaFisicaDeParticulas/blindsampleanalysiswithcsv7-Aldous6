```python
import numpy as np
import pandas as pd
data = pd.read_csv('/content/Sample7.csv')
```

```python
data.head()
```

```python
# Filtro los datos con ambas cargas positivas
ambas_cargas_positivas = data[(data['Q1'] == 1) & (data['Q2'] == 1)]

# Filtro los datos con ambas cargas negativas
ambas_cargas_negativas = data[(data['Q1'] == -1) & (data['Q2'] == -1)]

# Filtro los datos con una carga positiva y una negativa
Carga = data[(data['Q1'] == 1) & (data['Q2'] == -1) | (data['Q1'] == -1) & (data['Q2'] == 1)]

# Obtener el número de datos en cada categoría
num_ambas_cargas_positivas = len(ambas_cargas_positivas)
num_ambas_cargas_negativas = len(ambas_cargas_negativas)
num_una_positiva_una_negativa = len(Carga)

# Imprimir los resultados
print("Número de datos con ambas cargas positivas:", num_ambas_cargas_positivas)
print("Número de datos con ambas cargas negativas:", num_ambas_cargas_negativas)
print("Número de datos con una carga positiva y una negativa:", num_una_positiva_una_negativa)
```

```python
CargaSet = Carga.copy()
num_total_nuevo_dataframe = len(CargaSet)
print("Número total de datos en el nuevo DataFrame:", num_total_nuevo_dataframe)

```

```python
CargaSet
```

```python
def calcular_masa_invariante(pt1, eta1, phi1, pt2, eta2, phi2):
    return np.sqrt(2 * pt1 * pt2 * (np.cosh(eta1 - eta2) - np.cos(phi1 - phi2)))

# Aplicamos la función calcular_masa_invariante a cada fila de datos y agregar los resultados como una nueva columna 'masa_invariante'
CargaSet.loc[:, 'masa_invariante'] = calcular_masa_invariante(CargaSet['pt1'], CargaSet['eta1'], CargaSet['phi1'], CargaSet['pt2'], CargaSet['eta2'], CargaSet['phi2'])

print(CargaSet[['Run', 'Event', 'masa_invariante']].head())


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(CargaSet['masa_invariante'], bins=50, color='red', edgecolor='black')
plt.xlabel('Masa Invariante [GeV]')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Masa Invariante')
plt.grid(True)
plt.show()
```

```python
todos_tipos_cargaset = pd.concat([CargaSet['type1'], CargaSet['type2']])

tipos_totales_cargaset = todos_tipos_cargaset.unique()

print("Tipos totales encontrados en CargaSet:", tipos_totales_cargaset)

```

```python
# Calculamos la variable de aislamiento combinada (iso)
CargaSet['iso1'] = CargaSet['isoTrack1'] + CargaSet['isoEcal1'] + CargaSet['isoHcal1']
'
print("Primeros valores de la variable de aislamiento combinada (iso1):")
print(CargaSet['iso1'].head())

promedio_iso1_cargaset = CargaSet['iso1'].mean()

print("El promedio de la nueva variable 'iso1' en CargaSet es:", promedio_iso1_cargaset)
```

```python
CargaSet['iso2'] = CargaSet['isoTrack2'] + CargaSet['isoEcal2'] + CargaSet['isoHcal2']

print("Primeros valores de la variable de aislamiento combinada (iso2):")
print(CargaSet['iso2'].head())

promedio_iso2 = CargaSet['iso2'].mean()

print("El promedio de la nueva variable 'iso2' es:", promedio_iso2)
```

```python
import matplotlib.pyplot as plt

# Crear una figura y ejes
plt.figure(figsize=(10, 6))

# Graficar iso1
plt.plot(CargaSet.index, CargaSet['iso1'], label='iso1', color='blue')

# Graficar iso2
plt.plot(CargaSet.index, CargaSet['iso2'], label='iso2', color='red')

# Añadir etiquetas y título
plt.xlabel('Índice de muestra')
plt.ylabel('Valor de aislamiento')
plt.title('Valores de aislamiento iso1 e iso2')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
import plotly.express as px
CargaSet['iso1'] = CargaSet['isoTrack1'] + CargaSet['isoEcal1'] + CargaSet['isoHcal1']
CargaSet['iso2'] = CargaSet['isoTrack2'] + CargaSet['isoEcal2'] + CargaSet['isoHcal2']

fig = px.histogram(CargaSet, x=['iso1', 'iso2'], nbins=50, barmode='overlay', labels={'value': 'Valor de la variable de aislamiento', 'variable': 'Variable de aislamiento'}, color_discrete_sequence=['blue', 'red'])
fig.update_layout(title='Distribución de las variables de aislamiento combinadas', xaxis_title='Valor de la variable de aislamiento', yaxis_title='Frecuencia')
fig.show()

```
